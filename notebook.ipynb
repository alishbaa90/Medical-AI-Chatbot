```python
import os
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

DATA_PATH = os.getcwd()

def load_pdf_files(data_path):
    loader = DirectoryLoader(data_path, glob='*.pdf', loader_cls=PyPDFLoader)
    documents = loader.load()
    return documents

documents = load_pdf_files(DATA_PATH)
print("Loaded PDF pages:", len(documents))

def create_chunks(docs):
    splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    chunks = splitter.split_documents(docs)
    return chunks

text_chunks = create_chunks(documents)
print("Created text chunks:", len(text_chunks))
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

loader = PyPDFLoader("/content/The_GALE_ENCYCLOPEDIA_of_MEDICINE_SECOND.pdf")
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_chunks = text_splitter.split_documents(documents)

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.from_documents(text_chunks, embedding_model)

DB_FAISS_PATH = "/content/vectorstore/db_faiss"
db.save_local(DB_FAISS_PATH)
print("FAISS DB saved at:", DB_FAISS_PATH)

```

```python
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings

DB_FAISS_PATH = "/content/vectorstore/db_faiss"
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.load_local(DB_FAISS_PATH, embedding_model, allow_dangerous_deserialization=True)

retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})

llm = ChatGroq(
    api_key="your-api-key",
    model_name="llama3-8b-8192",
    temperature=0.2
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
while True:
    query = input("\nAsk your question (or type 'exit' to quit): ")
    if query.lower() == "exit":
        break

    result = qa_chain.invoke({'query': query})
    answer = result.get('result', '').strip()
    sources = result.get('source_documents', [])

    if not answer or not sources:
        print("\nAnswer: Sorry, I couldn’t find an answer related to your question in the uploaded PDF.")
    else:
        print("\nAnswer:", answer)
        print("\nSource Document(s):")
        for i, doc in enumerate(sources):
            print(f"\n--- Source {i+1} ---")
            print(doc.page_content)
```

```python
%%writefile app.py
import streamlit as st
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings

DB_FAISS_PATH = "/content/vectorstore/db_faiss"
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.load_local(DB_FAISS_PATH, embedding_model, allow_dangerous_deserialization=True)
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})

llm = ChatGroq(
    api_key="your-api-key",
    model_name="llama3-8b-8192",
    temperature=0.2
)

custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a helpful medical assistant. Use ONLY the following context to answer the question.
If the answer is not in the context, respond with "I don’t know based on the document."

Context:
{context}

Question: {question}
Answer:"""
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    chain_type_kwargs={"prompt": custom_prompt},
    return_source_documents=True
)

st.title("🩺 Medical AI Chatbot")

user_query = st.text_input("Ask your question:")
if user_query:
    result = qa_chain.invoke({'query': user_query})

    st.subheader("Answer")
    st.write(result["result"])

    st.subheader("Source Document")
    for doc in result["source_documents"]:
        st.write(doc.page_content[:300] + "...")
```
